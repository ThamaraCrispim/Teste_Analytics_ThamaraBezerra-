```python
#bibliotecas
import sqlite3
import pandas as pd
from google.colab import drive
```

```python
# 2. Carregar os dados
df = pd.read_csv('/content/drive/MyDrive/Dados/Dados.test/data_clean.csv')
```

```python
# 3. Criar banco SQLite em memória
conn = sqlite3.connect(':memory:')
df.to_sql('vendas', conn, index=False, if_exists='replace')
```

```python
# 4. Testar uma consulta (exemplo)
query = "SELECT * FROM vendas LIMIT 5;"
pd.read_sql(query, conn)
```

```python
# Consulta 1: Total de vendas por produto (ordenado do maior para o menor)
query1 = """
SELECT
    Produto,
    Categoria,
    SUM(Quantidade * Preço) AS Total_Vendas
FROM vendas
GROUP BY Produto, Categoria
ORDER BY Total_Vendas DESC;
"""

df_result1 = pd.read_sql(query1, conn)
print("Consulta 1 – Total de vendas por produto:")
display(df_result1)
```

Analise: Podemos ver aqui, assim como vimos no gráfico de barras, que o Tijolo Cerâmico lidera com mais da metade do faturamento, seguido por Tinta Acrílica e Furadeira. Isso confirma que ele tem alta demanda, mas qualquer problema com esse produto (falta de estoque, aumento de preço) pode impactar fortemente a receita da empresa.

```python
query2a = """
SELECT
    Produto,
    SUM(Quantidade * Preço) AS Total_Vendas_Junho
FROM vendas
WHERE strftime('%Y-%m', Data) = '2023-06'   -- Adaptado para 2023 (dados disponíveis)
GROUP BY Produto
ORDER BY Total_Vendas_Junho ASC
LIMIT 5;
"""

df_result2a = pd.read_sql(query2a, conn)
print("Consulta 2a – Top 5 produtos com menor faturamento em junho/2023:")
display(df_result2a)
```

Analise: No gráfico de barras (faturamento total do ano), esses mesmos produtos já apareciam como os que menos vendem no ano inteiro.
A consulta de junho mostra que, em um mês específico, eles continuam sendo os piores.
